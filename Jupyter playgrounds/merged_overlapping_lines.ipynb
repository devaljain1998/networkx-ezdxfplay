```python
import shapely
from shapely.ops import linemerge
from shapely.geometry import LineString, MultiLineString
```

```python
lines = [
    LineString([(0, 0), (1, 0)]),
    LineString([(2, 0), (1, 0)]),
    LineString([(2, 0), (3, 0)]),
    LineString([(4, 0), (3, 0)]),
]

multi_line = MultiLineString(lines)
```

```python
multi_line
```

```python
merged_line = linemerge(lines)
```

```python
print(merged_line)
```

```python
new_lines = [
    LineString([(2, 0), (5, 0)]),
    LineString([(0, 0), (10, 0)])
]
new_multi_line = MultiLineString(new_lines)
new_merged_line = linemerge(new_multi_line)
```

```python
merged_line(multi_line)
```

```python
merged_line = linemerge(multi_line)
```

```python
merged_line
```

```python
print(merged_line)
```

```python
print(new_merged_line)
```

```python
new_lines = [
    LineString([(2, 0), (5, 0)]),
    LineString([(0, 0), (10, 0)])
]
new_union_of_lines = shapely.ops.unary_union(new_lines)
new_multi_line = MultiLineString(new_lines)
new_merged_line = linemerge(new_multi_line)
new_merged_line1 = linemerge(new_union_of_lines)
```

```python
new_union_of_lines
```

```python
print(new_union_of_lines)
```
